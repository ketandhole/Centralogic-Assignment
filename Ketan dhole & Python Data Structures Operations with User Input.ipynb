```python
# ast stands for Abstract Syntax Trees.It imports the ast module, making its functions and classes available for use in the current Python script or session.

import ast

# User Input:- In this we take two inputs in which 1st will be decided it will be list, tuple, set or dictionary.The 2nd input
#                take the initial values for data types using the main function.

def main():
    data_structure_type = input("Enter the data structure type (list, tuple, set, dictionary): ").strip().lower()
    initial_values = input("Enter the initial values for the data structure: ")
    
# Input Parsing:
#              Using (ast.literal_eval) to safely parse the input string into the correct data structure.

    try:
        # Safely evaluate the input
        if data_structure_type == "dictionary":
            # Ensure the input is enclosed in curly braces
            initial_values = "{" + initial_values + "}"
        data = ast.literal_eval(initial_values)
    except Exception as e:
        print("Error in input values:", e)
        return
    
    if data_structure_type == "list":
        data = list(data)
        list_operations(data)
    elif data_structure_type == "tuple":
        data = tuple(data)
        tuple_operations(data)
    elif data_structure_type == "set":
        data = set(data)
        set_operations(data)
    elif data_structure_type == "dictionary":
        data = dict(data)
        dictionary_operations(data)
    else:
        print("Invalid data structure type.")

if __name__ == "__main__":
    main()

#List Operations:- Implementing a function & perform all the methods of List.

def list_operations(data):
    print("Initial list:", data)
    
    # append
    data.append(100)
    print("After append(100):", data)
    
    # extend
    data.extend([101, 102])
    print("After extend([101, 102]):", data)
    
    # insert
    data.insert(1, 200)
    print("After insert(1, 200):", data)
    
    # remove
    data.remove(200)
    print("After remove(200):", data)
    
    # pop
    data.pop()
    print("After pop():", data)
    
    # pop with index
    data.pop(0)
    print("After pop(0):", data)
    
    # clear
    data.clear()
    print("After clear():", data)
    
    # repopulate for further operations
    data.extend([3, 1, 2])
    print("After repopulating with [3, 1, 2]:", data)
    
    # sort
    data.sort()
    print("After sort():", data)
    
    # reverse
    data.reverse()
    print("After reverse():", data)
    
    # index
    index = data.index(1)
    print("Index of 1:", index)
    
    # count
    count = data.count(2)
    print("Count of 2:", count)
    
    # copy
    new_data = data.copy()
    print("Copy of list:", new_data)

# Tuple Operations:- Implementing a function & perform all the methods of Tuple.

def tuple_operations(data):
    print("Initial tuple:", data)
    
    # count
    count = data.count(data[0])
    print(f"Count of {data[0]}:", count)
    
    # index
    index = data.index(data[0])
    print(f"Index of {data[0]}:", index)

# Set operations:- Implementing a function & perform all the methods of Set. 

def set_operations(data):
    print("Initial set:", data)
    
    # add
    data.add(100)
    print("After add(100):", data)
    
    # update
    data.update([101, 102])
    print("After update([101, 102]):", data)
    
    # remove
    data.remove(100)
    print("After remove(100):", data)
    
    # discard
    data.discard(999)
    print("After discard(999):", data)
    
    # pop
    popped_value = data.pop()
    print("After pop():", data, "(popped:", popped_value, ")")
    
    # clear
    data.clear()
    print("After clear():", data)
    
    # repopulate for further operations
    data.update([1, 2, 3])
    print("After repopulating with {1, 2, 3}:", data)
    
    # union
    new_data = data.union({4, 5})
    print("Union with {4, 5}:", new_data)
    
    # intersection
    new_data = data.intersection({1, 2})
    print("Intersection with {1, 2}:", new_data)
    
    # difference
    new_data = data.difference({1})
    print("Difference with {1}:", new_data)
    
    # symmetric_difference
    new_data = data.symmetric_difference({2, 3})
    print("Symmetric difference with {2, 3}:", new_data)
    
    # issubset
    is_subset = data.issubset({1, 2, 3, 4})
    print("Is subset of {1, 2, 3, 4}:", is_subset)
    
    # issuperset
    is_superset = data.issuperset({1})
    print("Is superset of {1}:", is_superset)
    
    # copy
    new_data = data.copy()
    print("Copy of set:", new_data)

# Dictionary operations:- Implementing a function & perform all the methods of Dictionary.

def dictionary_operations(data):
    print("Initial dictionary:", data)
    
    # clear
    data.clear()
    print("After clear():", data)
    
    # repopulate for further operations
    data.update({'name': 'ketan', 'age': 23})
    print("After repopulating with {'name': 'ketan', 'age': 23}:", data)
    
    # copy
    new_data = data.copy()
    print("Copy of dictionary:", new_data)
    
    # fromkeys
    new_data = dict.fromkeys(['key1', 'key2'], 'value')
    print("Fromkeys with ['key1', 'key2'] and 'value':", new_data)
    
    # get
    value = data.get('name')
    print("Get value of 'name':", value)
    
    # items
    items = data.items()
    print("Items:", items)
    
    # keys
    keys = data.keys()
    print("Keys:", keys)
    
    # pop
    value = data.pop('name')
    print("After pop('name'):", data, "(removed:", value, ")")
    
    # repopulate for further operations
    data.update({'name': 'ketan', 'age': 23})
    
    # popitem
    key, value = data.popitem()
    print("After popitem():", data, "(removed:", (key, value), ")")
    
    # repopulate for further operations
    data.update({'name': 'ketan', 'age': 23})
    1
    # setdefault
    value = data.setdefault('address', 'Unknown')
    print("After setdefault('address', 'Unknown'):", data, "(returned:", value, ")")
    
    # update
    data.update({'city': 'Nashik'})
    print("After update({'city': 'Nashik'}):", data)
    
    # values
    values = data.values()
    print("Values:", values)


# Input examples: 
# >list:
#       Enter the data structure type (list, tuple, set, dictionary): list
#       Enter the initial values for the data structure: 10,20,30
    
# >tuple:
#        Enter the data structure type (list, tuple, set, dictionary): tuple
#        Enter the initial values for the data structure: 1, 2, 3

# >set:
#      Enter the data structure type (list, tuple, set, dictionary): set
#      Enter the initial values for the data structure: 12,56,45
        
# >dictionary:
#             Enter the data structure type (list, tuple, set, dictionary): dictionary
#             Enter the initial values for the data structure: 'name': 'ketan', 'age': 23, 'domain': 'python' (in the form of 'key':'value')
```
